O arquivo <font style="font-family: monaco">sales1000.csv</font> possui informações relativas à venda de produtos. As colunas abaixo são de particular importância para solução dos exercícios:
- *Country*: país da venda
- *Item Type*: tipo do produto vendido no país
- *Units Sold*: quantidade vendida

````python
import pandas as pd
import numpy as np

df = pd.read_csv('sales1000.csv')
````

### Exercício 2)
Considerando todas as vendas de um país, qual o país com o maior lucro total?

a) Cameroon<br>
b) Japan<br>
c) Australia<br>
d) Cuba

````python
# Solucao
df.groupby(by = 'Country', as_index=False).sum().sort_values(by = 'Total Profit', ascending=False)
````

### Exercício 3)
Quantos países vendem exatamente sete produtos distintos? 

a) 9<br>
b) 8<br>
c) 7<br>
d) 6

````python
#Solucao
exe3 = df[['Country', 'Item Type']].drop_duplicates().groupby(by='Country').count()
exe3[exe3['Item Type'] == 7].count()
````

### Exercício 4)
Qual o país com a maior diferença entre a maior quantidade e a menor quantidade de produtos vendidos? Por exemplo, 'Tonga' vende os seguintes quantidades de produtos:
```python
Household  8635
Beverages  9367
Meat       9633
Fruits     7124
Meat       3499
```
Portando, a diferença entre a maior e a menor quantidade de produtos vendidos é $$9633-3499=6134$$ Qual o país com a maior diferença?

a) Tonga<br>
b) Cape Verde<br>
c) Romania<br>
d) Cuba

````python
temp = df.groupby(by = ['Country', 'Item Type'], as_index=False).sum()[['Country', 'Item Type', 'Units Sold']].sort_values(by=['Country', 'Units Sold'], ascending=False)
temp[temp['Country'] == 'Tonga']
````

````python
# Solucao
exe4 = df.groupby(by = ['Country', 'Item Type'], as_index=False).sum()[['Country', 'Item Type', 'Units Sold']].groupby(by = ['Country'], as_index=False).agg(max_units=('Units Sold', np.max), min_units=('Units Sold', np.min))
exe4['diff'] = exe4['max_units'] - exe4['min_units']
exe4[exe4['Country'].isin(['Tonga', 'Cape Verde', 'Romania', 'Cuba'])].sort_values(by='diff', ascending=False).head(20)
````

### Exercício 5)
