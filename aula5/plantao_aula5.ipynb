# **Principais Gráficos Estatísticos:**

**Linhas:**
- plotado em um plano cartesiano para relacionar duas variáveis;
- utilizado para mostrar tendências de crescimento, decrescimento e estabilidade;
- repare no gráfico que as variáveis $y_1$ e $y_2$ possuem o valor $x$ em comum.

**Muito aplicado em gráficos de séries temporais e treinamento de redes neurais**

```python
import matplotlib.pyplot as plt
import numpy as np

x = [10, 20, 30, 40, 50, 60]
y1 = [1, 5, 3, 5, 7, 8]
y2 = [5, 7, 8, 6, 9, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

**Barras ou colunas:**
- utilizado para comparar frequências de uma única variável;
- um gráfico de colunas muito importante é o Histograma, que nos permite verificar a distribuição dos valores dentro do conjunto.

**Muito utilizado na análise dos atributos, como etapa de pré-processamento dos dados**

```python
print("Valor de faturamento na última semana:")
plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ('Fernando', 'Ana', 'Maria', 'José', 'Elisa')
y_pos = np.arange(len(people))
performance = 10000 * np.random.rand(len(people))

ax.barh(y_pos, performance, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()
ax.set_xlabel('Quantidade')
plt.show()

print("Valores:", performance)
print("Média:", performance.mean())
print("Desvio padrão:", performance.std())
print("Somatório:", performance.sum())
```

**Dispersão:**
- é uma representação de um par $(x,y)$ que não possui uma componente temporal;
- utilizado para verificar se há alguma correlação entre os exemplos.

**Correlação:**
- sem correlação: há apenas uma nuvem de pontos;
- correlação positiva forte: quando X aumenta, Y aumenta conjuntamente;
- correlação positiva média: quando X aumenta, Y tende a aumentar tambem;
- correlação negativa forte: quando X aumenta, Y diminui conjuntamente;
- correlação negativa média: quando X aumenta, Y tende a diminuir tambem.

**Muito utilizado em técnicas de agrupamento**

```python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]

plt.scatter(x, y)
plt.show()
```

**Boxplot:**
- ou diagrama de caixa permite visualizar a distribuição de valores discrepantes;
- sua formação depende da mediana e interquartis;
- valores que estejam fora do intervalo interquartil são considerados *outliers*:
  - Limite Inferior = Primeiro Quartil – 1,5 * (Terceiro Quartil – Primeiro Quartil)
  - Limite Superior = Terceiro Quartil + 1,5 * (Terceiro Quartil – Primeiro Quartil)

**Muito utilizado em análise de outliers**

```python
idades = np.array([16, 24, 29, 39, 42, 43, 45, 48, 49, 51, 60, 68, 75, 81, 82, 85])
print(idades)
print("Mediana:", np.median(idades))
print("25% Quartil:", np.percentile(idades, 25, axis=0))
print("75% Quartil:", np.percentile(idades, 75))
print("Interquartil:", np.percentile(idades, 75)-np.percentile(idades, 25))
print("Limite inferior:", np.percentile(idades, 25) - 1.5 * (np.percentile(idades, 75)-np.percentile(idades, 25)))
print("Limite superior:", np.percentile(idades, 75) + 1.5 * (np.percentile(idades, 75)-np.percentile(idades, 25)))

idades2 = np.array([16, 24, 29, 39, 42, 43, 45, 48, 49, 51, 60, 68, 75, 81, 82, 85, 900])
print("\n\n")
print(idades2)
print("Mediana:", np.median(idades2))
print("25% Quartil:", np.percentile(idades2, 25, axis=0))
print("75% Quartil:", np.percentile(idades2, 75))
print("Interquartil:", np.percentile(idades2, 75)-np.percentile(idades2, 25))
print("Limite inferior:", np.percentile(idades2, 25) - 1.5 * (np.percentile(idades2, 75)-np.percentile(idades2, 25)))
print("Limite superior:", np.percentile(idades2, 75) + 1.5 * (np.percentile(idades2, 75)-np.percentile(idades2, 25)))

plt.figure(figsize=(8,4))
plt.subplot(121); plt.boxplot(idades)
plt.subplot(122); plt.boxplot(idades2)
plt.show()
```
