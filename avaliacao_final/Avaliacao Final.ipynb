### Questão 4 (3 pontos)
Faça o gráfico da evolução do número acumulado de contaminados no Brasil (Brazil) e na Itália (Italy) ao longo do tempo. 

```python
import plotly.express as px
```

```python
df_melt = pd.melt(df, id_vars = ['Country/Region'], value_vars=df.columns[4:], var_name='date',value_name ='cases')
```

```python
df_melt[df_melt['Country/Region'] == 'Brazil']
```

```python
df4 = df_melt.groupby(by = ['Country/Region', 'date'], as_index=False).sum()
```

```python
df4 = df4[ df4['Country/Region'].isin(['Brazil', 'Italy']) ]
df4
```

```python
from datetime import datetime
```

```python
df4['date'] = df4['date'].apply(lambda x : datetime.strptime(x, '%m/%d/%y'))
```

```python
df4 = df4.reset_index(drop=True)
```

```python
df4.sort_values(by=['Country/Region', 'date'])
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
sns.lineplot(data=df4, x="date", y="cases", hue='Country/Region')
```
